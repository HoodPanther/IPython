```python
import numpy as np
import pandas as pd
import scipy.stats as st
import urllib2
from scipy.stats import norm
```

```python
def mk_test(x, alpha = 0.05):
    """
    this perform the MK (Mann-Kendall) test to check if there is any trend present in 
    data or not
    
    Input:
        x:   a vector of data
        alpha: significance level
    
    Output:
        trend: tells the trend (increasing, decreasing or no trend)
        h: True (if trend is present) or False (if trend is absence)
        p: p value of the sifnificance test
        z: normalized test statistics 
        
    Examples
    --------
      >>> x = np.random.rand(100)
      >>> trend,h,p,z = mk_test(x,0.05) 
      
    Credit: http://pydoc.net/Python/ambhas/0.4.0/ambhas.stats/
    """
    n = len(x)
    
    # calculate S 
    s = 0
    for k in xrange(n-1):
        for j in xrange(k+1,n):
            s += np.sign(x[j] - x[k])
    
    # calculate the unique data
    unique_x = np.unique(x)
    g = len(unique_x)
    
    # calculate the var(s)
    if n == g: # there is no tie
        var_s = (n*(n-1)*(2*n+5))/18
    else: # there are some ties in data
        tp = np.zeros(unique_x.shape)
        for i in xrange(len(unique_x)):
            tp[i] = sum(unique_x[i] == x)
        var_s = (n*(n-1)*(2*n+5) + np.sum(tp*(tp-1)*(2*tp+5)))/18
    
    if s>0:
        z = (s - 1)/np.sqrt(var_s)
    elif s == 0:
        z = 0
    elif s<0:
        z = (s + 1)/np.sqrt(var_s)
    else:
        z = 0
    
    # calculate the p_value
    p = 2*(1-norm.cdf(abs(z))) # two tail test
    h = abs(z) > norm.ppf(1-alpha/2) 
    
    if (z<0) and h:
        trend = 'decreasing'
    elif (z>0) and h:
        trend = 'increasing'
    else:
        trend = 'no trend'
        
    return trend, h, p, z, s 
```

```python
infile= r'http://nwis.waterdata.usgs.gov/usa/nwis/qwdata/?site_no=11530500&agency_cd=USGS&inventory_output=0&rdb_inventory_output=file&TZoutput=0&pm_cd_compare=Greater%20than&radio_parm_cds=parm_cd_list&radio_multiple_parm_cds=00665&qw_attributes=0&format=rdb&qw_sample_wide=wide&rdb_qw_attributes=0&date_format=YYYY-MM-DD&rdb_compression=value&submitted_form=brief_list'
response = urllib2.urlopen(infile)
cols = ['datetime','agency','site','end_dt','end_tm','dtm','dtm_cd','coll_ent_cd','medium_cd','tu_id','body_part_id','PO4']

usgs = pd.read_table(response, skiprows=65, skip_footer=5, na_values=('-','','No Data','No data','< 0.01','E 0.03','E 0.05','E 0.02'),parse_dates={'datetime':[2,3]})#, parse_cols=cols)
usgs.columns = cols

usgs.set_index('datetime',inplace=True)
usgs.drop(['agency','site','end_dt','end_tm','dtm','dtm_cd','coll_ent_cd','medium_cd','tu_id','body_part_id'],axis=1,inplace=True)
usgs['month'] = usgs.index.to_datetime().month.astype(int)
usgs['year'] = usgs.index.to_datetime().year.astype(int)
usgs = usgs[(usgs['year']>=1972) & (usgs['year']<=1980)]
usgs.sort_index(axis=0, inplace=True)

usgs.sort(columns=['month','year'],axis=0, inplace=True)
usgs.dropna(inplace=True)

usgsGrp = usgs.groupby(['month']).apply(lambda x: st.kendalltau(x['year'], x['PO4'],initial_lexsort=False)).to_frame()
usgsGrp.reset_index(inplace=True)
usgsGrp['p'] = usgsGrp.iloc[:,1].apply(lambda x: x[1],1)
usgsGrp['tau'] = usgsGrp.iloc[:,1].apply(lambda x: x[0],1)
usgsGrp.drop([0],axis=1,inplace=True)
print usgsGrp['tau'].sum()
print norm.pdf(usgsGrp['tau'].sum(),usgsGrp['tau'].mean(),usgsGrp['tau'].std())
print usgsGrp
```

```python
infile2 = r'http://pubs.usgs.gov/sir/2012/5049/support/SIR2012_5049_Appendix_2_04022012.xlsx'
response2 = urllib2.urlopen(infile2)
cols = ['site','unit','network','date','time','Nrem','N','Clrem','Cl','TDSrem', 'TDS']

usgsT = pd.read_excel(response2, skiprows=1, skip_footer=5, na_values=('-','','No Data','No data','.'))#, parse_cols=cols)

usgsT.columns = cols

usgsT['strTime'] = usgsT['time'].apply(lambda x: '0000' if np.isnan(x)==True else str(str(int(x)).zfill(4)))
usgsT['strdate'] = usgsT['date'].apply(lambda x: str(x)[:10])

usgsT['datetime'] = pd.to_datetime(usgsT['strdate'] + usgsT['strTime'], format='%Y-%m-%d%H%M', coerce=True)
usgsT.drop(['strTime','strdate'],axis=1,inplace=True)
usgsT.reset_index(inplace=True)
usgsT.set_index('datetime',inplace=True)
usgsT.sort_index(axis=0, inplace=True)

usgsT['year'] = usgsT.index.to_datetime().year.astype(int)
usgsT.sort(columns=['network', 'site', 'year'],axis=0, inplace=True)
```

```python
df = usgsT.loc[usgsT['Nrem']!='<']
df = df.loc[np.isfinite(df['N'])]

def taunow(x):
    return np.median(x['N']), len(x['N']), st.kendalltau(x['year'], x['N'])

def sqsig(x):
    return (x/18.0)*(x-1.0)*(2.0*x+5.0)

def detS(s,sigma):
    if s > 0:
        return (s-1)/sigma
    if s < 0:
        return (s+1)/sigma
    else:
        return 0

#usgsG1 = df.groupby(['network','site'])['N'].agg([np.median, np.average, np.size])
#usgsGroup = df.groupby(['network','site']).apply(lambda x: st.kendalltau(x['year'], x['N'])).to_frame()
usgsGroup = df.groupby(['network','site']).apply(lambda x: taunow(x)).to_frame()
#usgsGroup = pd.merge(usgsGroup,usgsG1,how='inner', left_index=True,right_index=True)
usgsGroup.reset_index(inplace=True)
#usgsGroup['conc'] = usgsGroup.iloc[:,2].apply(lambda x: x[0])
usgsGroup['n'] = usgsGroup.iloc[:,2].apply(lambda x: x[1])
usgsGroup['tau'] = usgsGroup.iloc[:,2].apply(lambda x: x[2][0])
#usgsGroup['pval'] = usgsGroup.iloc[:,2].apply(lambda x: x[2][1])
usgsGroup['sqsig'] = usgsGroup.loc[:,'n'].apply(lambda x: sqsig(x))
usgsGroupSum = usgsGroup.groupby(['network']).agg([np.sum, np.median, np.average, np.size, np.argmax]).reset_index()
#usgsGroupSum = usgsGroup.groupby(['network']).median()
usgsGroupSum.loc[:,('tau','sigma')] = np.sqrt(usgsGroupSum.loc[:,('sqsig','sum')])
usgsGroupSum.loc[:,('tau','z')] = usgsGroupSum[[('tau','median'),('tau','sigma')]].apply(lambda x: detS(x[0],x[1]),1)
usgsGroupSum.loc[:,('tau','pval')] =usgsGroupSum[[('tau','median'),('tau','sigma')]].apply(lambda x: st.norm.pdf(x[0],0,x[1]),1) 
usgsGroupSum
```

```python
#df = usgsT.loc[usgsT['Nrem']!='<']
#df = df.loc[np.isfinite(df['N'])]
df = usgs
const = 'PO4'
region = 'month'
block = ''
date = 'datetime'

df.reset_index(inplace=True)
df.set_index(date,inplace=True)
df['year'] = df.index.to_datetime().year.astype(int)
df['month'] = df.index.to_datetime().month.astype(int)
df['day'] = df.index.to_datetime().day.astype(int)
#
df.reset_index(inplace=True)

if block=='' or block==None:
    df.sort(columns=[region, date],axis=0, inplace=True)
    df.set_index([region,date],inplace=True)
else:
    df.sort(columns=[region, block, date],axis=0, inplace=True)
    df.set_index([region,block],inplace=True)

idx = pd.IndexSlice
banks = df.index.get_level_values(0).unique()

for i in banks:
    if block=='' or block==None:
        if len(df.loc[i,const])<3:
            pass
        else:
            df.loc[i,'diff'] = df.loc[i,const].diff()
            df.loc[idx[i,:],'p'] = df.loc[idx[i,:],'diff'].apply(lambda x: 1 if x>0 else 0).cumsum()
            df.loc[idx[i,:],'m'] = df.loc[idx[i,:],'diff'].apply(lambda x: 1 if x<0 else 0).cumsum()
            df.loc[idx[i,:],'t'] = df.loc[idx[i,:],'diff'].apply(lambda x: 1 if x==0 else 0).cumsum()
            df.loc[idx[i,:],'s'] = df.loc[idx[i,:],'p'] - df.loc[idx[i,:],'m']
            df.loc[idx[i,:],'n'] = np.abs(df.loc[idx[i,:],'p']) + np.abs(df.loc[idx[i,:],'m'])        
    else:
        # generate list of sites for each region
        sites = []
        sites = list(df.loc[idx[i,:]].index)
        for j in sites:
            if len(df.loc[(i,j),const])<3:
                pass
            else:
                df.loc[(i,j),'diff'] = df.loc[(i,j),const].diff()
                df.loc[idx[i,j],'p'] = df.loc[(i,j),'diff'].apply(lambda x: 1 if x>0 else 0).cumsum()
                df.loc[idx[i,j],'m'] = df.loc[(i,j),'diff'].apply(lambda x: 1 if x<0 else 0).cumsum()
                df.loc[idx[i,j],'t'] = df.loc[(i,j),'diff'].apply(lambda x: 1 if x==0 else 0).cumsum()
                df.loc[idx[i,j],'s'] = df.loc[(i,j),'p'] - df.loc[(i,j),'m']
                df.loc[idx[i,j],'n'] = np.abs(df.loc[(i,j),'p']) + np.abs(df.loc[(i,j),'m'])

if block=='' or block==None:
    summary = df.reset_index()
    summary = summary.groupby([region]).last()
    summary['tau'] = summary[['s','n','t']].apply(lambda x: x[0]/((x[1]*(x[1]-1))/2),1)
else:
    summary = df.reset_index()
    summary = summary.groupby([region,block]).last()
    summary['tau'] = summary[['s','n','t']].apply(lambda x: x[0]/((x[1]*(x[1]-1))/2),1)


#df.reset_index(inplace=True)
#df.set_index(['network'],inplace=True)
```
