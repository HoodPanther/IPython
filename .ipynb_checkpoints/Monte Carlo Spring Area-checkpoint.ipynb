# Optimization with scipy

```python
discharge_rng = (20000.0,35000.0)
thickness_rng = (500.0,8000.0)
eff_porosity_rng = (0.01,0.2)
age_rng = (7700.0, 16600.0)
delT_rng = (3.738,17.738)
heatFlux_rng = (39.0,155.0)
n = 1000
bnds = [discharge_rng, thickness_rng, eff_porosity_rng, age_rng, delT_rng, heatFlux_rng]
```

recharge rate=discharge/area <br>
Equation 2a <br>
area = (discharge*age)/(thickness*effective porosity)<br>
Equation 3a <br>
area = (discharge*(p*C*delT))/heat flux<br>

((Q*age)/(thick*poros)) - ((Q*(999.58*1.001*delT))/heatFlux)

```python
def area(x):
    arg1 = ((x[0]*x[3])/(x[1]*x[2]))
    arg2 = ((x[0]*(999.58*1.001*x[4]))/x[5])
    return np.abs(arg1-arg2)
```

```python
%%timeit
Q, thick, poros, age, delT, heatFlux = [],[],[],[],[],[]

for i in range(n):
    x = differential_evolution(area,bnds,maxiter=5000).x
    Q.append(x[0])
    thick.append(x[1])
    poros.append(x[2])
    age.append(x[3])
    delT.append(x[4])
    heatFlux.append(x[5])
```

```python
d = {'discharge':Q, 'thickness':thick, 'effective porosity':poros, 'age':age,
    'delT':delT, 'heat flux':heatFlux}
results = pd.DataFrame(data = d)
```

```python
results['area'] = results[['discharge','age','thickness','effective porosity']].apply(lambda x: (x[0]*x[1])/(x[2]*x[3]),1)
```

```python
results['area'].hist(bins=20)
plt.xlabel('Area (acres)')
plt.ylabel('Number of Occurrences')
```

# Optimization with SPOTPY

http://fb09-pasig.umwelt.uni-giessen.de/spotpy/Tutorial/5-Hydrological_model/

https://github.com/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/tree/master/

http://fb09-pasig.umwelt.uni-giessen.de/cmf/wiki/CmfDownload

```python
class spotpy_setup(object):
    def __init__(self):
        self.params = [spotpy.parameter.Uniform('discharge',20000.0,35000.0,1,25000.0),
                       spotpy.parameter.Uniform('thickness',500.0,8000.0,1,3000.0),
                       spotpy.parameter.Uniform('porosity',0.01,0.2,0.01,0.1),
                       spotpy.parameter.Uniform('age',7700.0,16600.0,5.0,9000.0),
                       spotpy.parameter.Uniform('delT',3.738,17.738,0.1,10.0),
                       spotpy.parameter.Uniform('flux',39.5,155.5,0.5,100.0)
                       ]
    
    def parameters(self):
        return spotpy.parameter.generate(self.params)
    
    def simulation(self,vector):
        simulations = [(((vector[0]*vector[3])/(vector[1]*vector[2])) - ((vector[0]*(999.58*1.001*vector[4]))/vector[5]))]
        return simulations

    def evaluation(self):
        observations=[0]
        return observations
    
    def objectivefunction(self,simulation,evaluation):
        objectivefunction= -spotpy.objectivefunctions.rmse(evaluation,simulation)      
        return objectivefunction
    
```

```python
spotpy_setup=spotpy_setup()
```

```python
results=[]

rep=5000

sampler=spotpy.algorithms.mc(spotpy_setup,    dbname='aMC',    dbformat='csv')
sampler.sample(rep)
results.append(sampler.getdata())

sampler=spotpy.algorithms.lhs(spotpy_setup,   dbname='aLHS',   dbformat='csv')
sampler.sample(rep)
results.append(sampler.getdata())

sampler=spotpy.algorithms.mle(spotpy_setup,   dbname='aMLE',   dbformat='csv')
sampler.sample(rep)
results.append(sampler.getdata())

sampler=spotpy.algorithms.mcmc(spotpy_setup,  dbname='aMCMC',  dbformat='csv') # Good for this
sampler.sample(rep)
results.append(sampler.getdata())

sampler=spotpy.algorithms.sceua(spotpy_setup, dbname='aSCEUA', dbformat='csv') # Good for this
sampler.sample(rep,ngs=2)
results.append(sampler.getdata())

sampler=spotpy.algorithms.sa(spotpy_setup,    dbname='aSA',    dbformat='csv')
sampler.sample(rep)
results.append(sampler.getdata())

sampler=spotpy.algorithms.demcz(spotpy_setup, dbname='aDEMCz', dbformat='csv') # Good for this
sampler.sample(rep,nChains=30)
results.append(sampler.getdata())

sampler=spotpy.algorithms.rope(spotpy_setup,  dbname='aROPE',  dbformat='csv')
sampler.sample(rep)
results.append(sampler.getdata())
```

```python
algDf= {}
algorithms=['MC','LHS','MLE','MCMC','SCEUA','SA','DEMCz','ROPE']
#algorithms=['MC','MCMC','SA']
pars = ['like','discharge','thickness','porosity','age','delT','flux','simulation','chain']
for i in range(len(algorithms)):
    algDf[algorithms[i]] = pd.DataFrame(data = results[i])
allAlgs = pd.concat(algDf)
allAlgs.reset_index(inplace=True)
allAlgs.set_index('level_1',inplace=True)
```
