```python
def smoother(df, p, win, sd):
    '''
    remove outliers from a pandas dataframe column and fill with interpolated values
    warning: this will fill all NaN values in the dataframe with the interpolate function
    df= dataframe of interest
    p= column in dataframe with outliers
    win= size of window
    std= number of standard deviations allowed
    '''
    df1 = df
    df1['dp'+ p] = df1[p].diff()
    df1['ma'+ p] = pd.rolling_mean(df1['dp'+ p], window=win, center=True)
    df1['mst'+p] = pd.rolling_std(df1['dp'+ p], window=win, center=True)
    for i in df.index:
        try:
            if abs(df1['dp'+ p][i] - df1['ma'+ p][i]) >= abs(df1['mst'+p][i]*sd):
                df[p][i]=np.nan
            else:
                df[p][i]=df[p][i]
        except (ValueError):
            try:
                if abs(df1['dp'+ p][i] - df1['ma'+ p][i]) >= abs(df1['dp'+p].std()*sd):
                    df[p][i]=np.nan
                else:
                    df[p][i]=df[p][i]
            except (ValueError):
                df[p][i]=df[p][i]
    del df1
    df = df.interpolate(method='time')
    df = df[1:-1]
    return df
```

```python
def rollmeandiff(df1,p1,df2,p2,win):
    '''
    returns the rolling mean difference of two columns from two different dataframes
    df1 = dataframe 1
    p1 = column in df1
    df2 = dataframe 2
    p2 =column in df2
    win = window in days
    '''
    win = win*60*24
    df1 = df1.resample('1Min', how='mean')
    df1 = df1.interpolate(method='time')
    df2 = df2.resample('1Min', how='mean')
    df2 = df2.interpolate(method='time')
    df1['rm'+p1] = pd.rolling_mean(df1[p1], window=win, center=True)
    df2['rm'+p2] = pd.rolling_mean(df2[p2], window=win, center=True)
    df3 = pd.merge(df1,df2,left_index=True, right_index=True, how='outer')
    df3 = df3[np.isfinite(df3['rm'+p1])]
    df4 = df3[np.isfinite(df3['rm'+p2])]
    df5 = df4['rm'+p1] - df4['rm'+p2]
    diff = round(df5.mean(),3)
    del (df3,df4,df5)
    return diff
```

#### Arnold Well Data Import

```python
arnold_well = pd.read_csv(arnwell, skiprows=15,parse_dates={'dt':[0,1]}, index_col='dt',engine='python')
arnold_well['water_above_trans_ft'] = arnold_well['LEVEL']
arnold_well['aw_temp_C'] = arnold_well['TEMPERATURE']
arnold_well = arnold_well.drop(['ms','LEVEL','TEMPERATURE'],axis=1)
arnold_well['water_above_trans_ft'].plot()
plt.xlabel('date')
plt.title('Arnold Well')
plt.ylabel('absolute pressure in feet of water above transducer')
plt.ylim(25,45)
```

```python
arnold_well['d_water_above_trans_ft'] = arnold_well['water_above_trans_ft'].diff(periods=3)
arnold_well_Jump = arnold_well[abs(arnold_well['d_water_above_trans_ft'])>5.0]
print arnold_well_Jump
```

```python
for i in arnold_well.index:
    if i < pd.datetime(2014,12,4,10):
        arnold_well['water_above_trans_ft'][i]=arnold_well['water_above_trans_ft'][i]
    elif i > pd.datetime(2014,12,4,11):
        arnold_well['water_above_trans_ft'][i] = arnold_well['water_above_trans_ft'][i]-arnold_well_Jump['d_water_above_trans_ft'][2]
    else:
        arnold_well['water_above_trans_ft'][i] = np.nan
```

```python
arnold_well['water_above_trans_ft'].fillna(inplace=True, method='FFILL')
arnold_well['water_above_trans_ft'].plot()
plt.xlabel('date')
plt.title('Arnold Well')
plt.ylabel('absolute pressure in feet of water above transducer')
plt.ylim(28,32)
```
